```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('Fish.csv')
dataset
```

```python
print("The Different Species are :",list(dataset['Species'].unique()))
```

```python
print("The data for the species Bream and Perch are :")
dataframe=pd.DataFrame(dataset[dataset['Species'].isin(['Bream','Perch'])])
dataframe.index=range(len(dataframe))
dataframe
```

```python
dataframe.isnull()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
scaler=StandardScaler()
dataframe.iloc[:,1:]=scaler.fit_transform(dataframe.iloc[:,1:])
xtrain,xtest,ytrain,ytest=train_test_split(dataframe.iloc[:,1:].values,dataframe['Species'].values,test_size=0.15)
```

### Scikit Learn Implementation

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model=KNeighborsClassifier()
knn_model.fit(xtrain,ytrain)
```

```python
print("Training R2 score : ",knn_model.score(xtrain,ytrain))
```

```python
ypred=knn_model.predict(xtest)
print("Predictions \n",ypred)
```

```python
print("True Values \n",ytest)
```

```python
conf_mat=confusion_matrix(ytest,ypred)
print("Confusion Matrix \n",conf_mat)
print("\nClassification Report : \n",classification_report(ytest,ypred))
print("Accuracy : " ,(conf_mat[0][0]+conf_mat[1][1])/len(ytest) )
```

```python
probs=knn_model.predict_proba(xtest)
probs=probs[:,1]
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelBinarizer
```

```python
binarizer=LabelBinarizer()
ytest_b=binarizer.fit_transform(ytest)
fpr,tpr,_=roc_curve(ytest_b,probs)
random_probs = [0 for _ in range(len(ytest))]
p_fpr,p_tpr,_ = roc_curve(ytest_b,random_probs)
auc_score=roc_auc_score(ytest,probs)
print("AUC SCORE : " ,auc_score)
```

```python
plt.plot(p_fpr, p_tpr, linestyle='--')
plt.plot(fpr, tpr, marker='.', label='KNN Classifier (area=%0.2f)'% auc_score)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("ROC-AUC CURVE for KNN Classifier")
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
ytrain_b=binarizer.fit_transform(ytrain).flatten()
```

```python
from sklearn.metrics import scorer
```

```python
knn = KNeighborsClassifier()
params = {'n_neighbors':[2,5,7],'metric':['minkowski','manhattan','euclidean']}
grid_search = GridSearchCV(knn,param_grid = params,scoring='precision',cv=2)
grid_search.fit(xtrain,ytrain_b)
print(grid_search.best_params_)
print(grid_search.best_score_)
```

```python
grid_search.cv_results_
```

```python
knn = KNeighborsClassifier()
params = {'n_neighbors':np.arange(2,11,step=1),'metric':['minkowski','manhattan','euclidean']}
random_search = RandomizedSearchCV(knn,params,scoring='precision' ,cv=3)
random_search.fit(xtrain,ytrain_b)
```

```python
print(random_search.best_params_)
print(random_search.best_score_)
```
